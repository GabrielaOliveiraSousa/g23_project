```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv(r"C:\Users\axlea\OneDrive\1ºano\g23_project\Jupyter\g23_Podcasts_Guests.csv", sep=';')
df.head()
```

```python
conn = sqlite3.connect("DadaBase_Podcast.db")
```

```python
guests = df[['guest_id', 'name']].drop_duplicates()
guests.to_sql('Guest', conn, if_exists='replace', index=False)
```

```python
podcasts = df[['podcast_id', 'title', 'category', 'creation_date']].drop_duplicates()
podcasts.to_sql('Podcast', conn, if_exists='replace', index=False)
```

```python
participation = df[['guest_id', 'podcast_id', 'date', 'amount']]
participation.to_sql('Participation', conn, if_exists='replace', index=False)
```

```python
sponsors = df[['sponsor_id', 'podcast_id', 'extra_info']].dropna().drop_duplicates()
sponsors.to_sql('Sponsor', conn, if_exists='replace', index=False)
```

```python
themes = df[['podcast_id', 'subject']].drop_duplicates()
themes['theme_id'] = range(1, len(themes) + 1)
themes.to_sql('Theme', conn, if_exists='replace', index=False)
```
